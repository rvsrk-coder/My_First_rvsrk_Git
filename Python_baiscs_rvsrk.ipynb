ask 1: Variables and Data Types Create variables to store a person's name (string), age (integer), height in meters (float), and whether they are a student (boolean). Print all four values in a single print() statement.

```python
name = "John"
age = 20
height = 1.75
is_student = True

print(f"name =  {name} age = {age}, height = { height}, is_student={is_student}")
print("Datya types ....")
print(f"name data type : {type(name)}")
print(f"age data type : {type(age)}")
print(f"height data type : {type(height)}")
print(f"is_student data type : {type(is_student)}")
```

Task 2: String Formatting Using the variables from Task 1, print a sentence in the following format: Name: <name> | Age: <age> | Height: <height> | Student: <is_student>

```python
print(f"Name: {name} | Age: {age} | Height: {height} | Student: {is_student}")
```

Task 3: Arithmetic Operations Using the age variable, calculate and print:

Age in months
Age in days (assume 365 days/year)
The remainder when age is divided by 7
Age raised to the power of 2

```python
# Age calculations
age_in_months = age * 12
age_in_days = age * 365
remainder = age % 7
age_square = age ** 2

print("Age in months:", age_in_months)
print("Age in days:", age_in_days)
print("Remainder when divided by 7:", remainder)
print("Age squared:", age_square)
```
